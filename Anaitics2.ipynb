```python
import geopy.geocoders
from geopy.geocoders import Nominatim 
geopy.geocoders.options.default_timeout = 7
def localiazar(nombre):  
    loc = Nominatim(user_agent="GetLoc") 
    
    getLoc = loc.geocode(query=nombre) 
    
    if getLoc is not None:
        if getLoc.address is not None:
            lat= getLoc.latitude
            long= getLoc.longitude
        else:
            lat = long = np.NaN
        sleep(2)
        
    else:
        lat = long = np.NaN
    return lat, long
```

```python
#funcion para convertir consumo miles de barriles a twh
def calculos_petroleo(df):
    #print('en funcion petrole')
    #df=df.astype({'annual_production':'int64'})
    #p=10**(-9)
    #eq_xj=6.11786319999928*p
    twh=588.4407483972509
    for i, ele in enumerate(df['Consumo_Petroleo']): 
        totaltwh=df.loc[i,'Consumo_Petroleo']*twh
        df.loc[i,'Consumo_Petroleo']=totaltwh
    return(df)
```

```python
#funcion para convertir produciion miles de barriles a exajoules
def calculos_petroleo_prod(df):
    #print('en funcion petrole')
    #df=df.astype({'annual_production':'int64'})
    #p=10**(-9)
    #eq_xj=6.11786319999928*p
    exa=0.0000000612
    for i, ele in enumerate(df['Produccion_Petroleo']): 
        totalexa=df.loc[i,'Produccion_Petroleo']*exa
        df.loc[i,'Produccion_Petroleo']=totalexa
    return(df)
```

```python
df_pet_tw=calculos_petroleo(ene_pet)
df_pet_tw
```

```python
df_pet_tw=calculos_petroleo(ene_pet)
df_pet_tw=calculos_petroleo_prod(df_pet_tw)
df_pet_tw
```

```python
petroleo80=df_pet_tw.groupby(('Id_anio')).Consumo_Petroleo.mean()
petroleo80=df_pet_tw[df_pet_tw['Anio']>=1980]
petroleo80.round({'Consumo_Petroleo':2,'Produccion_Petroleo':2})

petroleo80
```
