```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
energia = pd.read_csv('D:\data\Proyectos\Grupal\Grupal-DTS02\proyecto_grupal_consumo_energ-a_co2\energyco2.csv')
energia = energia.drop(columns='Unnamed: 0')
energia_mayor = energia.sort_values(by='CO2_emission',ascending=False)
energia_mayor = energia_mayor[energia_mayor['Country']!= 'World']
emergia_mayor_2019 = energia_mayor[energia_mayor['Year'] == 2019]
"""
    Paises con mayor emision de CO2 en el año 2019 (año de ultimo registro obtenido)
"""
colores=['blueviolet','deeppink','mediumslateblue','deepskyblue','cyan']
emergia_mayor_2019.groupby('Country')['CO2_emission'].sum().sort_values(ascending=False).head().plot(kind='bar',color = colores,figsize=(20,10))
plt.title('Top 5 paises con mayor emision de CO2 en 2019',fontsize=30)
plt.show()
```
